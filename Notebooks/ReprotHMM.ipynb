## Fixation Entropy

By checking the fixation entropy we can understand if there are particular AOIs that are more fixated than others. This can be useful to understand if there are particular areas of the screen that are more interesting to the user.

This helps us also to understand if we have probability of obtaining intersting results using the HMM method.

```python
# calculate entropy of the fixation distribution using stats.entropy
def EntropyFix(fixation,sizeX,sizeY):
    # fixation = np.array([x,y]...) array of fixations
    # create a 2D histogram of the fixations
    fixHist, xedges, yedges = np.histogram2d(fixation[:,0],fixation[:,1],bins=[sizeX,sizeY])
    # normalize the histogram
    fixHist=fixHist/np.sum(fixHist)
    # calculate entropy
    entropy=stats.entropy(fixHist.flatten())
    return entropy
```

```python
entropy = []
for stim in UrbanDat.stimuli:
    fixation = np.empty((0, 2))
    for s in UrbanDat.subjects:
        X,Y = UrbanDat.GetFixationData(s,stim)
        fixation = np.vstack((fixation, np.array([X, Y]).T))
    
    entropy.append(EntropyFix(fixation,1280,960))

# calculate mean and std of the entropy
mean_entropy = np.mean(entropy)
std_entropy = np.std(entropy)
print(mean_entropy,std_entropy)
```

With increasing of fixation number we start to have very high fitting time for the models.

Also with increasing of fixation entropy we start to have models components that fit to very large areas of the screen.

It's very difficult in this way to find an optimal number of components, and also to have a clear understanding of the areas that are more fixated by the subjects.

We could think to try to fit one component for each word that is present in the image, this as a reasonable task to give to the HMM, but as we will show later this can cause errors in the fitting of the model.

In conclusion we can say that HMM sill a valid method to use for this type of data, but when we use it is important to remember that:
 - if we have a high fixation entropy we could have problems in finding the optimal number of states
 - finding the optimal number of states using BIC, expecially with high number of fixations, could be very time consuming
 - with high number of components the results could be not very clear and can also changhe a lot from one run to another
